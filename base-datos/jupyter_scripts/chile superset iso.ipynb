```python
chile_input = 'chile.geojson'
chile_output = 'new_chile.geojson'
```

```python
chile = gpd.GeoDataFrame.from_file(chile_input)
chile['ISO'] = chile_data['iso']

chile.insert(loc=1, column='ID_0', value=48, allow_duplicates= True)


chile.pop('objectid')
chile.insert(loc=1, column='NAME_0', value='Chile', allow_duplicates= True)
chile = chile.rename(columns={'codregion': 'ID_1'})


chile = chile.rename(columns={'Region': 'NAME_1'})
chile['NAME_1'] = chile_data['name']

chile['TYPE_1'] = 'Región'
chile['ENGTYPE_1'] = 'Region'
chile['NL_NAME_1'] = None
chile['VARNAME_1'] = None


chile.pop('layer')
chile.pop('st_length_')
chile.pop('st_area_sh')
chile.pop('area_km')
chile.pop('cir_sena')
chile.pop('path')       

chile.to_file(chile_output, driver='GeoJSON')

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(chile)

#AGREGAR COMUNAS ACÁ
```
