```python
import csv
import re
import string
import json
import numpy as np
from ast import literal_eval

import openpyxl

file_needs = 'needs_v2.xlsx'

output_file_pair_exp_need = 'need_exp_pairs.csv'
output_file_pair_role_need = 'need_role_pairs.csv'

out_file_pair_exp = 'pair_exp.csv'
out_file_pair_roles = 'pair_roles.csv'


pair_need_columnas = [
    'need_id',
    'pair_id',
]

pair_words_columnas = [
    'pair_id',
    'palabra_inicio',
    'palabra_fin',
]
```

```python
book = openpyxl.load_workbook(file_needs)

sheet = book.active

pair_needs_role = list()
pair_words_role = list()

pair_needs_exp = list()
pair_words_exp = list()


pair_ids_exp = 0
pair_ids_role = 0

for row in sheet.iter_rows(min_row=2):
    need_id = int(row[8].value)
    
    role_token = row[6].value
    if role_token == None or role_token == 'NR':
        role_token = '[]'
        
    array_role_token = np.array(literal_eval(role_token))    
    
    pairs_role = []
    
    for i in range(0,len(array_role_token) - 1):
        pairs_role.append((array_role_token[i], array_role_token[i+1]))
        
    for pair in pairs_role:
        pair_words_role.append({
            'pair_id': pair_ids_role,
            'palabra_inicio': pair[0],
            'palabra_fin': pair[1],
        })
        
        pair_needs_role.append({
            'pair_id': pair_ids_role,
            'need_id': need_id,
        })
        
        pair_ids_role += 1        
        
    exp_token = row[7].value
    if exp_token == None or exp_token == 'NR':
        exp_token = '[]'
        
    array_exp_token = np.array(literal_eval(exp_token))    
    
    pairs_exp = []
    
    for i in range(0,len(array_exp_token) - 1):
        pairs_exp.append((array_exp_token[i], array_exp_token[i+1]))      
    
    for pair in pairs_exp:
        pair_words_exp.append({
            'pair_id': pair_ids_exp,
            'palabra_inicio': pair[0],
            'palabra_fin': pair[1],
        })
        
        pair_needs_exp.append({
            'pair_id': pair_ids_exp,
            'need_id': need_id,
        })
        
        pair_ids_exp += 1     

        
with open(output_file_pair_role_need, 'w') as new_file:
    writer = csv.DictWriter(new_file, fieldnames = pair_need_columnas)
    writer.writeheader()
    for data in pair_needs_role:
        writer.writerow(data)
        
with open(out_file_pair_roles, 'w') as new_file:
    writer = csv.DictWriter(new_file, fieldnames = pair_words_columnas)
    writer.writeheader()
    for data in pair_words_role:
        writer.writerow(data)        
        
with open(output_file_pair_exp_need, 'w') as new_file:
    writer = csv.DictWriter(new_file, fieldnames = pair_need_columnas)
    writer.writeheader()
    for data in pair_needs_exp:
        writer.writerow(data)
        
with open(out_file_pair_exp, 'w') as new_file:
    writer = csv.DictWriter(new_file, fieldnames = pair_words_columnas)
    writer.writeheader()
    for data in pair_words_exp:
        writer.writerow(data)                 
```
