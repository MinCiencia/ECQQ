```python
import csv
import re
import string
import jsonz

import openpyxl

file_needs = 'needs.xlsx'
output_file = 'need_data.sql'

values = list()
```

```python
book = openpyxl.load_workbook(file_needs)

sheet = book.active

values = list()

name_len = 0
actor_len = 0
role_len = 0
name_long = ''
actor_long = ''
role_long = ''

for row in sheet.iter_rows(min_row=2):
    need_id = int(row[0].value)
    
    name = str(row[1].value)
    if name != None:
        name_len_aux = len(name)
        if name_len_aux > name_len:
            name_len = name_len_aux
            name_long = name
        if name != 'NR':
            name = name.replace('\'','')
            name = name.capitalize()
        if name == '\'':
            name = 'NR'    
    
    actor = str(row[2].value)
    if actor != None:
        actor_len_aux = len(actor)
        if actor_len_aux > actor_len:
            actor_len = actor_len_aux
            actor_long = actor
        if actor != 'NR':
            actor = actor.capitalize()       
                
    
    role = str(row[3].value)
    if role != None:
        role_len_aux = len(role)
        if role_len_aux > role_len:
            role_len = role_len_aux
            role_long = role
        if role != 'NR':
            role = role.replace('\'','')
            role = role.capitalize()
        if role == '\'':
            role = 'NR'
            
    role_token = row[4].value
    if role_token == None or role_token == 'NR':
        role_token = '[]'
        
    macro = str(row[5].value)  
            
    if macro != None:
        if macro == '\'':
            macro = 'NR'
        if macro != 'NR':
            macro = macro.replace('\'','')    
    
    if actor == 'State':
        actor = 'Estado'
    
    string_value = '''({},\'{}\',\'{}\',\'{}\',\'{}\',\'{}\')'''.format(need_id, name, actor, role, role_token.replace('\'','"'), macro)
    values.append(string_value)

print(name_len)
print(name_long)
print(actor_len)
print(actor_long)
print(role_len)
print(role_long)
        
with open(output_file, 'w') as new_file:
    for index, value in enumerate(values):
        if index == 0:
            print('INSERT INTO need VALUES {},'.format(value), file=new_file)
        elif index == len(values) - 1:
            print('''{};'''.format(value), file=new_file)
        else:
            print('{},'.format(value), file=new_file)
```
