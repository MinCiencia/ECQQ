```python
import csv
import re
import string

file_comunas = 'comunas.csv'
file_dialogos = 'BD/CSV/dialogue.csv'
output_file = 'dialogue.csv'

comunas = dict()

new_columnas = [
    'file_id', 
    'date',
    'init_time',
    'end_time',
    'location',
    'address',
    'comuna',
    'n_members',
    'group_name',
]
```

```python
with open(file_comunas) as csv_file:
    csv_reader = csv.DictReader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:       
        id_com = row['id']
        nombre = row['name']         
        comunas[nombre] = id_com   
        line_count += 1
    

with open(file_dialogos) as csv_file:
    csv_reader = csv.DictReader(csv_file, delimiter=',')
    new_dialogos = list()
    for row in csv_reader: 
        file_id = row['file_id']
        date = row['date']
        init_time = row['init_time']
        end_time = row['end_time']
        location = row['location']
        address = row['address']
        nom_comuna = row['comuna']
        if nom_comuna == '':
            nom_comuna = 'NR'
            comuna = comunas[nom_comuna]
        elif nom_comuna == 'Coihaique':
            nom_comuna = 'Coyhaique'
            comuna = comunas[nom_comuna]
        elif nom_comuna == 'Los Angeles':
            nom_comuna = 'Los Ángeles'
            comuna = comunas[nom_comuna]
        elif nom_comuna == 'Aisén':
            nom_comuna = 'Aysén'
            comuna = comunas[nom_comuna]
        elif nom_comuna == 'Ohiggins':
            nom_comuna = "O'higgins"
            comuna = comunas[nom_comuna]    
        elif nom_comuna == 'Ollague':
            nom_comuna = 'Ollagüe'
            comuna = comunas[nom_comuna]      
        elif nom_comuna == 'Paihuano':
            nom_comuna = 'Paiguano'
            comuna = comunas[nom_comuna]        
        elif nom_comuna == 'Llay Llay':
            nom_comuna = 'Llaillay'
            comuna = comunas[nom_comuna]   
        elif nom_comuna == 'Trehuaco':
            nom_comuna = 'Treguaco'
            comuna = comunas[nom_comuna]
        elif nom_comuna == 'Ranquil':
            nom_comuna = 'Ránquil'
            comuna = comunas[nom_comuna] 
        elif nom_comuna == 'Los Alamos':
            nom_comuna = 'Los Álamos'
            comuna = comunas[nom_comuna]
        elif nom_comuna == 'Til til':
            comuna = comunas['Tiltil']
        elif nom_comuna == 'Tirua':
            comuna = comunas['Tirúa']   
        elif nom_comuna == 'Requinoa':
            comuna = comunas['Requínoa']      
        elif nom_comuna == 'Hualaihue':
            comuna = comunas['Hualaihué']        
                 
        else:
            comuna = comunas[string.capwords(nom_comuna)]
            
        n_members = row['n_members']
        group_name = row['group_name']
            
        new_dialogos.append({
            'file_id' : file_id,
            'date' : date.replace('-','/'),
            'init_time' : init_time,
            'end_time' : end_time,
            'location' : location,
            'address' : address,
            'comuna' : comuna,
            'n_members' : n_members,
            'group_name' : group_name,
        })
                        
        line_count += 1
    
    try:
        with open(output_file, 'w') as new_csv_file:
            writer = csv.DictWriter(new_csv_file, fieldnames = new_columnas)
            writer.writeheader()
            for data in new_dialogos:
                writer.writerow(data)
    except IOError:
        print("I/O error")
```
