```python
import csv
import re
import string
import json

import openpyxl

file_needs = 'needs_v2.xlsx'
output_file = 'need_data.sql'

values = list()
```

```python
book = openpyxl.load_workbook(file_needs)

sheet = book.active

values = list()

for row in sheet.iter_rows(min_row=2):
    need_id = int(row[8].value)
    
    name = str(row[1].value)
    if name != None:
        if name != 'NR':
            name = name.replace('\'','')
            name = name.capitalize()
        if name == '\'':
            name = 'NR'    
    
    actor = str(row[2].value)
    if actor != None:
        if actor != 'NR':
            actor = actor.capitalize()  
                
    explanation = str(row[4].value)
    if explanation != None:
        if explanation != 'NR':
            explanation = explanation.replace('\'','')
            explanation = explanation.capitalize()
        if explanation == '\'':
            explanation = 'NR'
            
    role = str(row[3].value)
    if role != None:
        if role != 'NR':
            role = role.replace('\'','')
            role = role.capitalize()
        if role == '\'':
            role = 'NR'        
            
    role_token = row[6].value
    if role_token == None or role_token == 'NR':
        role_token = '[]'
        
    exp_token = row[7].value
    if exp_token == None or exp_token == 'NR':
        exp_token = '[]'    
        
    priority = int(row[5].value)    
        
    macro = str(row[9].value)  
            
    if macro != None:
        if macro == '\'':
            macro = 'NR'
        if macro != 'NR':
            macro = macro.replace('\'','')    
    
    if actor == 'State':
        actor = 'Estado'
    
    string_value = '''({},\'{}\',\'{}\',\'{}\',\'{}\',{},\'{}\',\'{}\',\'{}\')'''.format(
        need_id, 
        name, 
        actor, 
        role, 
        explanation, 
        priority, 
        role_token.replace('\'','"'), 
        exp_token.replace('\'','"'), 
        macro)
    values.append(string_value)
    
with open(output_file, 'w') as new_file:
    for index, value in enumerate(values):
        if index == 0:
            print('INSERT INTO need VALUES {},'.format(value), file=new_file)
        elif index == len(values) - 1:
            print('''{};'''.format(value), file=new_file)
        else:
            print('{},'.format(value), file=new_file)
```
